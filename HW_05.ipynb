```python
# Категориальные признаки
cat_cols = ['Genus', 'Species']

# Целевые значения
n_label = {'Leptodactylidae': df['Family'] == 'Leptodactylidae',
            'Hylidae': df['Family'] == 'Hylidae',
            'Dendrobatidae': df['Family'] == 'Dendrobatidae',
            'Bufonidae': df['Family'] == 'Bufonidae'}
```

**1. Нормализовать данные**

```python
# посмотрим на распределение числовых признаков
fig, axes = plt.subplots(7, 3, figsize=(20,30))
axes = axes.ravel()

for i, col in enumerate(num_cols[1:22]):
    sns.histplot(ax=axes[i], data=df, x=col, kde=True)
```

```python
# Посмотрим на распределение целевого признака
sns.countplot(x='Family', data=df);
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder, PolynomialFeatures
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
```

```python
def to_(df, name_target):
    
    X = df.drop(df.columns[-4:], axis=1)
    y = n_label[name_target].astype('int')
    
    scaler = StandardScaler()
    scaler.fit(X.values.reshape(-1, 22))
    X = scaler.transform(X.values.reshape(-1, 22))
    
    return X, y
```

**2. Построить бинарный классификатор для каждого семейства(Family) на основе LogisticRegression c $l_1$ и $l_2$ регуляризацими. Это соответствует целевым исходам (Конкретное семейство, Всё остальное). Подобрать гиперпараметр $C$ с помощью кросс-валидации, используя метрику ROC AUC.**

```python
# построим первую модель без учета гиперпараметров
for col in df['Family'].unique():
    data = to_(df, col)

    X_train, X_test, y_train, y_test = train_test_split(data[0], data[1], test_size=0.3, random_state=17)
    
    logit = LogisticRegression(penalty='l2')
    logit.fit(X_train, y_train)
    y_train_predicted = logit.predict_proba(X_train)[:, 1]
    y_test_predicted = logit.predict_proba(X_test)[:, 1]
    
    train_auc = roc_auc_score(y_train, y_train_predicted)
    test_auc = roc_auc_score(y_test, y_test_predicted)
    
    plt.figure(figsize=(10,7))
    plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
    plt.plot(*roc_curve(y_test, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
    legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
    legend_box.set_facecolor("white")
    legend_box.set_edgecolor("black")
    plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
    plt.title(col)
    plt.show()
```

```python
# С помощью кросс-валидации подберем наилучший параметр С
for col in df['Family'].unique():
    data = to_(df, col)

    X_train, X_test, y_train, y_test = train_test_split(data[0], data[1], test_size=0.3, random_state=17)
    
    logit = LogisticRegression(penalty='l2')
    param = {'C': [100, 10, 1, 0.1, 0.01, 0.001]}
    
    model_CV = GridSearchCV(logit, param, cv=5, n_jobs=-1, scoring='roc_auc')
    model_CV.fit(X_train, y_train)
    
    best_score = model_CV.best_score_
    best_param = model_CV.best_params_
    
    print(f'{col}: {best_score}, {best_param}')
```

```python
# обучим модель с наилучшими параметрами
for col in df['Family'].unique():
    data = to_(df, col)

    X_train, X_test, y_train, y_test = train_test_split(data[0], data[1], test_size=0.3, random_state=17)
    
    if col == 'Bufonidae':
        logit = LogisticRegression(C=0.1)
    else:
        logit = LogisticRegression(C=100)
    
    logit.fit(X_train, y_train)
    y_train_predicted = logit.predict_proba(X_train)[:, 1]
    y_test_predicted = logit.predict_proba(X_test)[:, 1]
    
    train_auc = roc_auc_score(y_train, y_train_predicted)
    test_auc = roc_auc_score(y_test, y_test_predicted)
    
    plt.figure(figsize=(10,7))
    plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
    plt.plot(*roc_curve(y_test, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
    legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
    legend_box.set_facecolor("white")
    legend_box.set_edgecolor("black")
    plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
    plt.title(col)
    plt.show()
```

**3. Добавить полиномиальные признаки до второго порядка. Построить бинарный классификатор для каждого семейства на основе LogisticRegression c $l_1$ и $l_2$ регуляризацими. Подобрать гиперпараметр $C$ с помощью кросс-валидации, используя метрику ROC AUC.**

```python
# Добавим полиномиальные признаки и обучим модель
for col in df['Family'].unique():
    data = to_(df, col)
    
    poly = PolynomialFeatures(2)
    poly_data = poly.fit_transform(data[0])

    X_train, X_test, y_train, y_test = train_test_split(poly_data, data[1], test_size=0.3, random_state=17)
    
    logit = LogisticRegression(penalty='l2')
    logit.fit(X_train, y_train)
    y_train_predicted = logit.predict_proba(X_train)[:, 1]
    y_test_predicted = logit.predict_proba(X_test)[:, 1]
    
    train_auc = roc_auc_score(y_train, y_train_predicted)
    test_auc = roc_auc_score(y_test, y_test_predicted)
    
    plt.figure(figsize=(10,7))
    plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
    plt.plot(*roc_curve(y_test, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
    legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
    legend_box.set_facecolor("white")
    legend_box.set_edgecolor("black")
    plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
    plt.title(col)
    plt.show()
```

```python
# С помощью кросс-валидации подберем наилучший параметр С
for col in df['Family'].unique():
    data = to_(df, col)
    
    poly = PolynomialFeatures(2)
    poly_data = poly.fit_transform(data[0])

    X_train, X_test, y_train, y_test = train_test_split(poly_data, data[1], test_size=0.3, random_state=17)
    
    logit = LogisticRegression(penalty='l2')
    param = {'C': [100, 10, 1, 0.1, 0.01, 0.001]}
    
    model_CV = GridSearchCV(logit, param, cv=5, n_jobs=-1, scoring='roc_auc')
    model_CV.fit(X_train, y_train)
    
    best_score = model_CV.best_score_
    best_param = model_CV.best_params_
    
    print(f'{col}: {best_score}, {best_param}')
```

```python
# обучим модель с наилучшими параметрами
for col in df['Family'].unique():
    data = to_(df, col)
    
    poly = PolynomialFeatures(2)
    poly_data = poly.fit_transform(data[0])

    X_train, X_test, y_train, y_test = train_test_split(poly_data, data[1], test_size=0.3, random_state=17)
    
    if col == 'Hylidae' or 'Leptodactylidae':
        logit = LogisticRegression(C=0.1)
    elif col == 'Dendrobatidae':
        logit = LogisticRegression(C=1)
    else:
        logit = LogisticRegression(C=0.01)
    
    logit.fit(X_train, y_train)
    y_train_predicted = logit.predict_proba(X_train)[:, 1]
    y_test_predicted = logit.predict_proba(X_test)[:, 1]
    
    train_auc = roc_auc_score(y_train, y_train_predicted)
    test_auc = roc_auc_score(y_test, y_test_predicted)
    
    plt.figure(figsize=(10,7))
    plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
    plt.plot(*roc_curve(y_test, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
    legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
    legend_box.set_facecolor("white")
    legend_box.set_edgecolor("black")
    plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
    plt.title(col)
    plt.show()
```

Наилучшие результаты показала модель с добавлением полиномиальных признаков, кроме последнего значения.
